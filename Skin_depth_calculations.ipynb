```python
import matplotlib.pyplot as plt
import numpy as np
from cmath import sqrt
from mpl_toolkits import mplot3d

STEEL = []
COPPER = []
ALUMINIUM = []

omega = np.arange(100000, 1000000000, 10000)


def skindepth(fr, mu, sigma):
    sd = sqrt(2/(fr*mu*sigma))
    return sd

for i in omega:
    st = skindepth(i, 0.000942, 1320000)
    STEEL.append(st)
    al = skindepth(i, 0.0000012566, 36900000)
    ALUMINIUM.append(al)
    cu = skindepth(i, 0.0000012566, 58700000)
    COPPER.append(cu)

plt.subplots()
plt.figure(figsize=(10,6))
plt.plot(omega, STEEL , 'b', label = 'Stainless Steel')
plt.plot(omega, ALUMINIUM , 'g', label = 'Aluminium')
plt.plot(omega, COPPER , 'r', label = 'Copper')
plt.xscale('log')
plt.xlabel('Frequency/Hz')
plt.ylabel('Skin Depth/m')
plt.legend()
plt.show()
```
